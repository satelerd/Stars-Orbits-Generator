```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def elipse(a,b,c, offset, star):
    plt.figure(figsize=(10,6))
    t = np.linspace(0,360,360)
    
    for i in range(0,3):
        x = a[i]*np.cos(np.radians(t)) + offset[i]
        y = b[i]*np.sin(np.radians(t))
        plt.plot(x,y)
    plt.plot(c[1],0,'ro')
    plt.legend(star)
    plt.grid()
    plt.show()
```

```python
a = [919, 1680, 1720]
b = [457.55, 386.2317957, 951.268101]
c = [-919*0.867247008-840, -1680*0.973214286+10, -1720*0.833139535-180]
offset = [-840,0,-180]
star = ['s02', 's016', 's019', 'Sgr A*']

elipse(a, b, c, offset, star)
```
