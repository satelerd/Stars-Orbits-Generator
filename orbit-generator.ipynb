```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def elipse(a,b,c, offset, star):
    plt.figure(figsize=(10,6))
    t = np.linspace(0,360,360)
    
    for i in range(0,3):
        x = a[i]*np.cos(np.radians(t)) + offset[i]
        y = b[i]*np.sin(np.radians(t))
        plt.plot(x,y)
    plt.plot(c[1],0,'ro')
    plt.legend(star)
    plt.grid()
    plt.show()
```

```python
# [S0-16, S0-19, S62]
a = [2219, 2312, 703.6063]
b = [495.016, 1089.96, 153.225]
c = [-2219*0.9748, -2312*0.8819, -703.6063*0.976]
offset = [110,0, -1380]
star = ['S0-16', 'S0-19', 'S62', 'Sgr A*']

elipse(a, b, c, offset, star)
```
